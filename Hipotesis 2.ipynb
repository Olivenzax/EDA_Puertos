```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mercancias_nav = pd.read_csv('data/Mercancias.csv',delimiter=";")

mercancias_nav
```

```python
mercancias_nav = mercancias_nav[mercancias_nav['Transito'] != "Solo tránsito"]
mercancias_nav = mercancias_nav[mercancias_nav['Autoridad Portuaria'] != "C.A.G.P."]

mercancias_nav
```

```python
mercancias_nav['Autoridad Portuaria'].value_counts()
```

```python
mercancias_nav.info()
```

```python
mercancias_nav = pd.DataFrame(mercancias_nav)

mercancias_nav
```

```python
mercancias_nav = mercancias_nav.loc[:,["Año", "Autoridad Portuaria", "Naturaleza", "TOTAL mercancías por naturaleza"]]
mercancias_nav
```

```python
mercancias_nav['TOTAL mercancías por naturaleza'] = mercancias_nav['TOTAL mercancías por naturaleza'].str.replace('.', '', regex=False).astype(int)

mercancias_nav.info()
```

```python
# Agrupar por 'Año' y 'Naturaleza', sumando el total de mercancías
total_por_naturaleza = mercancias_nav.groupby(['Año', 'Naturaleza'])['TOTAL mercancías por naturaleza'].sum().reset_index()

# Mostrar el resultado
total_por_naturaleza
```

```python
import pandas as pd

# Diccionario de mapeo (Naturaleza -> Sector)
mapeo_sectores = {
    # Energía
    "Petróleo crudo": "Energía",
    "Gas natural": "Energía",
    "Fueloil": "Energía",
    "Gasoil": "Energía",
    "Gasolina": "Energía",
    "Gases energéticos de petróleo": "Energía",
    "Biocombustibles": "Energía",
    "Carbón y coque de petróleo": "Energía",
    
    # Agroalimentario
    "Cereales y sus harinas": "Agroalimentario",
    "Frutas, hortalizas y legumbres": "Agroalimentario",
    "Vinos, bebidas, alcoholes y derivados": "Agroalimentario",
    "Conservas": "Agroalimentario",
    "Pescado congelado y refrigerado": "Agroalimentario",
    "Piensos y forrajes": "Agroalimentario",
    "Tabaco, cacao, café y especias": "Agroalimentario",
    "Habas de soja": "Agroalimentario",
    "Otros productos alimenticios": "Agroalimentario",
    
    # Minerales y Metales
    "Mineral de hierro": "Minerales y Metales",
    "Fosfatos": "Minerales y Metales",
    "Potasas": "Minerales y Metales",
    "Sal común": "Minerales y Metales",
    "Otros minerales no metálicos": "Minerales y Metales",
    "Otros minerales y residuos metálicos": "Minerales y Metales",
    "Chatarra de hierro": "Minerales y Metales",
    "Productos siderúrgicos": "Minerales y Metales",
    
    # Químicos e Industria
    "Productos químicos": "Químicos e Industria",
    "Abonos naturales y artificiales": "Químicos e Industria",
    "Cemento y clínker": "Químicos e Industria",
    "Papel y pasta": "Químicos e Industria",
    "Asfalto": "Químicos e Industria",
    "Materiales de construcción elaborados": "Químicos e Industria",
    
    # Manufacturas y Bienes de Equipo
    "Maquinaria, aparatos, herramientas y repuestos": "Manufacturas y Bienes de Equipo",
    "Vehículos y sus piezas": "Manufacturas y Bienes de Equipo",
    "Otros productos metalúrgicos": "Manufacturas y Bienes de Equipo",
    "Maderas y corcho": "Manufacturas y Bienes de Equipo",
    
    # Logística y Transporte
    "Taras de contenedores": "Logística y Transporte",
    "Taras de equipamiento ro-ro": "Logística y Transporte",
    "Mercancía en contenedores en tránsito": "Logística y Transporte",
    
    # Otros
    "Resto de mercancías": "Otros"
}

# Aplicar el mapeo al DataFrame
total_por_naturaleza['Sector'] = total_por_naturaleza['Naturaleza'].map(mapeo_sectores)

# Agrupar por Sector y sumar el total
df_agrupado = total_por_naturaleza.groupby(['Año', 'Sector'])['TOTAL mercancías por naturaleza'].sum().reset_index()

df_agrupado
```

```python
import seaborn as sns 

# Configurar estilo
sns.set_theme(style="whitegrid")

# Crear figura
plt.figure(figsize=(12, 6))

# Gráfico de líneas
sns.lineplot(
    data=df_agrupado,
    x="Año",
    y="TOTAL mercancías por naturaleza",
    hue="Sector",
    marker="o",
    linewidth=2.5,
)

# Personalizar
plt.title("Evolución del tráfico marítimo en España por tipo de mercancía (1962-2023)", fontsize=14)
plt.xlabel("Año")
plt.ylabel("Toneladas x 1e8")
plt.xticks(rotation=45)  # Rotar años para mejor lectura
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Leyenda fuera del gráfico

# Mostrar
plt.tight_layout()
plt.show()
```

```python
df_agrupado['Porcentaje'] = df_agrupado.groupby('Año')['TOTAL mercancías por naturaleza'].transform(lambda x: x/x.sum()*100).round(2)

df_agrupado
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado df
# Asegúrate de que "Porcentaje" esté en formato numérico
df_agrupado['Porcentaje'] = pd.to_numeric(df_agrupado['Porcentaje'], errors='coerce')

# Creamos una tabla dinámica con los años como columnas y los sectores como filas
pivot_df = df_agrupado.pivot_table(index='Sector', columns='Año', values='Porcentaje', aggfunc='sum')

# Transponemos para que los años estén en el eje X
pivot_df = pivot_df.T

# Hacemos el gráfico de columnas apiladas
pivot_df.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='berlin')

plt.title('Distribución porcentual por sector y año')
plt.ylabel('Porcentaje')
plt.xlabel('Año')
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
por_puerto = mercancias_nav[mercancias_nav['Autoridad Portuaria'] == 'Bilbao'] 

por_puerto['TOTAL mercancías por naturaleza'] = por_puerto['TOTAL mercancías por naturaleza'].str.replace('.', '', regex=False).astype(int)

por_puerto_pt = por_puerto.pivot_table(
    index='Año',
    columns='Naturaleza',
    values='TOTAL mercancías por naturaleza',
    aggfunc='sum'  # Por si hay duplicados
).reset_index()

por_puerto_pt
```

```python
# Verificar valores nulos en el DataFrame
print(por_puerto_pt.isnull().sum())

# Rellenar NaN con 0 (si es apropiado para tu análisis)
por_puerto_pt = por_puerto_pt.fillna(0)

# O eliminar filas con NaN (si prefieres)
por_puerto_pt = por_puerto_pt.dropna()
```

```python
año_seleccionado = 2023  # Cambia al año que desees
df_año = por_puerto_pt[por_puerto_pt["Año"] == año_seleccionado]
```

```python
datos_pie = df_año.drop(columns=["Año"]).iloc[0]
```

```python
plt.figure(figsize=(10, 10))  # Tamaño del gráfico

# Pie chart
plt.pie(
    datos_pie,
    labels=datos_pie.index,  # Nombres de las mercancías
    autopct='%1.1f%%',       # Porcentajes con 1 decimal
    startangle=90,           # Ángulo inicial (arriba)
    colors=sns.color_palette("viridis", len(datos_pie)),  # Paleta de colores
    wedgeprops={'edgecolor': 'white', 'linewidth': 1},    # Bordes blancos
    textprops={'fontsize': 7}                            # Tamaño de fuente
)

# Título
plt.title(f"Distribución de mercancías en {año_seleccionado}\nAutoridad Portuaria de Bilbao", 
          fontsize=16, pad=20)

# Mostrar
plt.tight_layout()
plt.show()
```
